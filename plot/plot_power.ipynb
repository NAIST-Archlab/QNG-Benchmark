```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import constant
# Data for computer A
data_a = {
    'n': np.array([2, 3, 4, 5, 6, 7, 8, 9]),
    'GD': [(28.5, 1), (29.7, 1), (29.1, 1), (28.9, 1), (29.5, 1), (29.0, 1), (30.0, 28), (30.0, 28)],
    'Adam': [(28.7, 1), (30.0, 1), (30.0, 1), (29.1, 1), (29.0, 1), (29.0, 1), (30.0, 28), (29.9, 28)],
    'QNG-QFSM-GD': [(48.0, 1), (48.4, 1), (36.0, 12), (33.0, 15), (32.4, 18), (29.3, 3), (31.0, 28), (30.0, 28)],
    'QNG-QFIM-GD': [(27.6, 1), (29.1, 1), (29.7, 1), (29.1, 1), (29.0, 1), (29.0, 1), (29.0, 28), (30.1, 28)],
    'QNG-QFIM-Adam': [(27.8, 1), (29.1, 1), (28.6, 1), (29.2, 1), (29.4, 1), (29.0, 1), (29.5, 28), (29.9, 28)]
}

# Data for computer B
data_b = {
    'n': np.array([2, 3, 4, 5, 6, 7, 8, 9]),
    'GD': [(39.5, 1), (41.0, 1), (41.0, 1), (38.8, 1), (40.0, 1), (40.0, 1), (50.0, 32), (70.0, 32)],
    'Adam': [(38.5, 1), (39.0, 1), (39.0, 1), (39.0, 1), (39.0, 1), (40.0, 1), (50.0, 32), (70.0, 32)],
    'QNG-QFSM-GD': [(109.0, 6), (102.0, 9), (60.0, 12), (55.5, 15), (49.0, 32), (40.0, 32), (50.0, 32), (70.0, 32)],
    'QNG-QFIM-GD': [(39.0, 1), (39.0, 1), (39.0, 1), (39.0, 1), (38.8, 1), (39.0, 1), (50.0, 32), (70.0, 32)],
    'QNG-QFIM-Adam': [(39.0, 1), (39.0, 1), (39.1, 1), (39.0, 1), (39.5, 1), (39.0, 1), (50.0, 32), (70.0, 32)]
}

# Function to plot data for a single computer
def plot_data(ax, data, title):
    markers = ['o', '^', 's', 'p', '*']
    
    for j, optimizer in enumerate(['GD', 'Adam', 'QNG-QFSM-GD', 'QNG-QFIM-GD', 'QNG-QFIM-Adam']):
        values = data[optimizer]
        power = np.array([v[0] for v in values])
        cores = np.array([v[1] for v in values])
        ax.plot(data['n'], cores, power, marker='o', color=constant.colorss[j], label=constant.optimizers_name[j])

    ax.set_xlabel(r'$n$')
    ax.set_ylabel('# Cores')
    ax.set_zlabel('Power (W)')
    ax.set_title(title)

# Create subplots
fig = plt.figure(figsize=(14, 6))

# Subplot for Computer A
ax1 = fig.add_subplot(121, projection='3d')
plot_data(ax1, data_a, 'Intel i9-10940X')

# Subplot for Computer B
ax2 = fig.add_subplot(122, projection='3d')
plot_data(ax2, data_b, 'Intel i9-13900K')

plt.legend(ncol = 5)
plt.tight_layout()
plt.show()
```
