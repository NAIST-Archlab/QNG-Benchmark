```python
import matplotlib.pyplot as plt
import numpy as np
labels = ['GD', 'Adam', 'QNG-Fubini-Study-GD', 'QNG-QFIM-GD', 'QNG-QFIM-Adam']
colors = ['violet', 'red', 'green', 'blue', 'orange']

fig, axs = plt.subplots(nrows = 2, ncols = 5, figsize=(20, 5))
j = 0
for num_qubit in range(2,8):
    row = j // len(labels)
    col = j % len(labels)
    loss = np.loadtxt(f"loss_basic_values_{num_qubit}.txt")
    for i in range(len(labels)):
        axs[row, col].plot(loss[i], color = colors[i], label = labels[i], alpha = 0.5)
    axs[row, col].set_title(rf"$n={num_qubit}$")
    axs[row, col].set_xticks([0, 10, 20, 30, 40, 50])
    if row == 0:
        axs[row, col].set_xticklabels([])
    if row == 1:
        axs[row, col].set_xlabel("# Iteration")
    if col == 0:
        axs[row, col].set_ylabel(r"$\mathcal{C} (\boldsymbol{\theta})$")
    else:
        axs[row, col].set_yticklabels([])
    axs[row, col].set_ylim(-0.05, 1.05)
    
    j += 1
plt.tight_layout()
plt.legend()
    
    
```
